```python
#Klib is a Python library used for importing, cleaning, analyzing and pre-processing data.
#By using this library we can speed up the process of cleaning and exploration in  Data Analysis.
```

```python
#importing the libraries
import klib
import seaborn as sns 
```

```python
#Using the Titanic dataset already provided by seaborn.
#Titanic dataset has many missing values and also has different categorical features which are helpfull.
df = sns.load_dataset('titanic') 
```

```python
df.head()
```

```python
#Exploring the dataset with various Klib functions for visualizing.
#1- klib.corr_mat(df) # returns a color-encoded correlation matrix.
klib.corr_mat(df)
```

```python
#Here Black color shows +ve correlation.
#Red one show -ve correlation.
```

```python
#2- klib.corr_plot(df) # returns a color-encoded heatmap, ideal for correlations
klib.corr_plot(df)
```

```python
#3- klib.dist_plot(df) # returns a distribution plot for every numeric feature
klib.dist_plot(df)
```

```python
#It return probability density distribution
```

```python
#4- klib.missingval_plot(df) # returns a figure containing information about missing values
klib.missingval_plot(df)
```

```python
#By just looking at the figure above we can visualize that Age(177) and Deck(688) has missing values.
```

```python
df.info()
```

```python
#We can see different datatypes over here which are maybe taking unneccessary memory.
```

```python
#Functions for leaning the dataset --
#1- klib.data_cleaning(df) # performs datacleaning (drop duplicates & empty rows/cols, adjust dtypes,...)
df_clean = klib.data_cleaning(df)
```

```python
df_clean.info()
```

```python
#Now the data types are also changed accordingly so they occupy lesser space
```

```python
#2- klib.drop_missing(df) # drops missing values, also called in data_cleaning()
klib.drop_missing(df)
```
